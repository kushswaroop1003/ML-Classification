# Over Sampling - SMOTE Method

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .25, random_state=101)
```

```python
x_train.shape
```

```python
from imblearn.over_sampling import SMOTE
ros = SMOTE()
x_ros, y_ros = ros.fit_resample(x_train, y_train)
```

```python
x1 = x_ros
y1 = y_ros
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x1.shape)
print(y1.shape)
```

# Various Models Implemenation and Evaluation

# 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x1,y1)
```

```python
y_pred_train = LR.predict(x1)
y_pred_test = LR.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y1, y_pred_train))
print("#######"*25)
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(classification_report(y1, y_pred_train))
print("########"*25)
print(classification_report(y_test, y_pred_test))
```

```python
# Finding: Since its a case of Online Purchase, precision and F-Score are the most important metrics.
# Precision Average is 88% and f1 Score average is 92%[Both for tests]
```
